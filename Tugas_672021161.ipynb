**2. Permen, Pie Chart.**

```python
import matplotlib.pyplot as plt

nama_permen = ['Mentos', 'Kopiko', 'Golia', 'Yupie', 'Fisherman']
Warna = ('#3DB7E4','#6B4C35','#D4AF37','#FF8849','#607573')
jumlah_permen = [52, 39, 78, 13, 78]
Explode = (0, 0.1, 0, 0, 0)

total_permen = sum(jumlah_permen)

peluang_kopiko = (jumlah_permen[nama_permen.index('Kopiko')] / total_permen) * 100

plt.figure(figsize=(6, 6))
plt.pie(
    jumlah_permen, labels=nama_permen,
    autopct='%1.2f%%',
    startangle=90,
    explode = Explode,
    shadow = True,
    colors = Warna,
    )
plt.title('Persentase Peluang Memilih Permen Kopiko')
plt.show()

print(f"Peluang memilih permen Kopiko: {peluang_kopiko:.2f}%")
```
